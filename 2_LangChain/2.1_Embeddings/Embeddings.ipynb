```python
#Importing libraries 
import os 
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
#Loading OpenaI Embedding
from langchain_openai import OpenAIEmbeddings

#Creating embedding object
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
embeddings
```

```python
#Converting a given text into Openai embedding
text="this is a tutorial for OpenAI Embedding"

#Result of embeddings
query_result = embeddings.embed_query(text)
query_result
```

```python
#Length of the embedding vector
len(query_result)
```

```python
#Importing library 
from langchain_openai import OpenAIEmbeddings

#Creating an Embedding object with dimensions to control output dimension 
embeddings = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)

#Converting a given text into embedding
text="This is a tutorial on OPENAI embedding"
query_result = embeddings.embed_query(text)
len(query_result)   
```

```python
#Importing the library 
from langchain_community.document_loaders import TextLoader

#Creating Loader instance and load the input file 
loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
#Here we perform text splitting
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
#Previewing the data 
final_documents[0].page_content
```

```python
embeddings.embed_query(final_documents[0].page_content)
```
