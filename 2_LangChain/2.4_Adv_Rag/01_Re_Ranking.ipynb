# RAG Re-Ranking


```python
#Defining the source document 
documents = [
    "This is a list which containing sample documents.",
    "Keywords are important for keyword-based search.",
    "Document analysis involves extracting keywords.",
    "Keyword-based search relies on sparse embeddings.",
    "Understanding document structure aids in keyword extraction.",
    "Efficient keyword extraction enhances search accuracy.",
    "Semantic similarity improves document retrieval performance.",
    "Machine learning algorithms can optimize keyword extraction methods."
]
```

```python
#length of the document
len(documents)
```

```python
#Use Sentence Transformer 
from sentence_transformers import SentenceTransformer
#load the model 
model_name = 'sentence-transformers/paraphrase-xlm-r-multilingual-v1'
#Build the model 
model = SentenceTransformer(model_name)
```

```python
#Embedding the input document
document_embeddings = model.encode(documents)
len(document_embeddings)
```

```python
#Checking the first value
document_embeddings[0]
```

```python
#Rechecking the dimension of the embedding
len(document_embeddings[0])
```

```python
#Modifying the output format in a document format 
for i, embedding in enumerate(document_embeddings):
    print(f"Document {i+1} embedding: {embedding}")
```

```python
#Defining the query 
query = "Natural language processing techniques enhance keyword extraction efficiency."

#Performing Embedding on the query 
query_embedding = model.encode(query)

print("Query Embedding:", query_embedding)
```

```python
#Performing Cosine Simillarity between Document Embedding and input query 
import numpy as np 
from sklearn.metrics.pairwise import cosine_similarity

#Perform the simillarity operation
simillarities = cosine_similarity(np.array([query_embedding]), document_embeddings)
```

```python
#Checking the simillarity score 
simillarities
```

```python
#Identifying the most simillar record
most_simillar_index = np.argmax(simillarities)
print(most_simillar_index)
```

```python
#Displaying the most simillar record 
most_simillar_document = documents[most_simillar_index]
print(most_simillar_document)
```

```python
#similarity score associated to above 
simillarities[0][most_simillar_index]
```

```python
#Arranging the index in thr order of simillarity in desc 
sorted_indices = np.argsort(simillarities[0])[::-1]
sorted_indices
```

```python
#ranked documents
ranked_docs = [ (documents[i], simillarities[0][i]) for i in sorted_indices]
ranked_docs
```

```python
print("Ranked Documents: ")

for rank, (document, simillarity) in enumerate(ranked_docs, start=1):
    print(f"Rank {rank}: Document - {document}, Similarity Score - {simillarity}")
```

```python
print("Top 4 documents")
for rank, (document, simillarity) in enumerate(ranked_docs[:4], start=1):
    print(f"Rank {rank}: Document - {document}, Similarity Score - {simillarity}")
```

BM25Okapi, often simply referred to as BM25, is a ranking function used by search engines and information retrieval systems to estimate the relevance of documents to a given search query.

```python
#Importing BM25 
from rank_bm25 import BM25Okapi

top_4_documents = [ doc[0] for doc in ranked_docs[:4]]

print(top_4_documents)
```

```python
#Tokenizing the document 
tokenized_top_4_documents = [doc.split() for doc in top_4_documents]
print(tokenized_top_4_documents)
```

```python
#Tokenising the input query 
tokenized_query = query.split()
print(tokenized_query)
```

```python
#Instantiating BM25
bm25 = BM25Okapi(tokenized_top_4_documents)
print(bm25)
```

```python
#Getting BM25 score
bm25_score = bm25.get_scores(tokenized_query)
print(bm25_score)
```

```python
#Sorting the indices based on arg value 
sorted_indices2 = np.argsort(bm25_score)[::-1]
print(sorted_indices2)
```

```python
#Original top 4 documents were
top_4_documents
```

```python
#Reranked documents
reranked_documents = [(top_4_documents[i], bm25_score[i]) for i in sorted_indices2]
print(reranked_documents)
```

```python
print("Rerank of top 4 Documents:")
for rank, (document, similarity) in enumerate(reranked_documents, start=1):
    print(f"Rank {rank}: Document - '{document}', Similarity Score - {similarity}")
```

# RAG + ReRanking using BM25

```python
# ── 1. imports ──────────────────────────────────────────────────────
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain import hub

import faiss
from rank_bm25 import BM25Okapi
import re, numpy as np

from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableLambda
```

```python
# ── 2. load & chunk the PDF ─────────────────────────────────────────
FILE_PATH = "llama2.pdf"
loader    = PyPDFLoader(FILE_PATH)
pages     = loader.load()

splitter  = RecursiveCharacterTextSplitter(
    chunk_size    = 500,
    chunk_overlap = 50
)
docs = splitter.split_documents(pages)
```

```python
# ── 3. build FAISS vector store (MiniLM embeddings) ────────────────
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

dim   = len(embeddings.embed_query("dummy text"))   # 384 for MiniLM
index = faiss.IndexFlatIP(dim)

vector_store = FAISS(
    embedding_function   = embeddings,
    index                = index,
    docstore             = InMemoryDocstore(),
    index_to_docstore_id = {}
)
vector_store.add_documents(docs)
```

```python
# ── 4. dense recall retriever (larger pool) ────────────────────────
initial_k       = 20
dense_retriever = vector_store.as_retriever(search_kwargs={"k": initial_k})
```

```python
# ── 5. BM-25 re-ranker ─────────────────────────────────────────────
def bm25_rerank_fn(inputs: dict, top_k: int = 5):
    """inputs = {"question": str, "docs": List[Document]}"""
    question, candidate_docs = inputs["question"], inputs["docs"]

    tokenize = lambda txt: re.findall(r"\w+", txt.lower())
    corpus   = [tokenize(d.page_content) for d in candidate_docs]

    bm25     = BM25Okapi(corpus)
    scores   = bm25.get_scores(tokenize(question))

    best_idx = np.argsort(scores)[::-1][:top_k]
    return [candidate_docs[i] for i in best_idx]

bm25_reranker = RunnableLambda(bm25_rerank_fn)
```

```python
# ── 6. prompt, model, helpers ──────────────────────────────────────
prompt     = hub.pull("rlm/rag-prompt")           # classic RAG prompt
model      = ChatGoogleGenerativeAI(model="gemini-1.5-flash")
out_parser = StrOutputParser()

def format_docs(doc_list):
    return "\n\n".join(d.page_content for d in doc_list)
```

```python
# ── 7. the full RAG chain ──────────────────────────────────────────
rag_chain = (
    {
        "docs":     dense_retriever,        # dense recall
        "question": RunnablePassthrough()
    }
    | bm25_reranker                        # lexical re-rank
    | (lambda d: {"context": format_docs(d), "question": ""})
    | prompt
    | model
    | out_parser
)
```

```python
# ── 8. test run ────────────────────────────────────────────────────
if __name__ == "__main__":
    query  = "What is a Llama model?"
    answer = rag_chain.invoke(query)
    print(f"Q: {query}\nA: {answer}")
```
